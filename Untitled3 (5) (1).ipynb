```python
import numpy as np

import pandas as pd

from tabulate import tabulate

from sklearn.metrics import mean_squared_error, r2_score

from sklearn.neighbors import KNeighborsRegressor

from sklearn. tree import DecisionTreeRegressor

from sklearn.svm import SVR

import xgboost as xgb
from xgboost import XGBRegressor

from sklearn.ensemble import RandomForestRegressor

from sklearn. linear_model import LinearRegression

regressors = {
'Linear Regression': LinearRegression(),
'K-Nearest Neighbors': KNeighborsRegressor(),
'Decision Tree': DecisionTreeRegressor(),
'Support Vector Machine': SVR(),
'Random Forest': RandomForestRegressor(random_state=42),
'xgb' : xgb.XGBRegressor (objective='reg:squarederror', random_state=42)
}
result_dict = {
'Regressor': [],
'Train_MSE' : [],
'Test_MSE' : [],
# 'Train_RMSE': [],
# 'Test_RMSE': [],
'Train_R2_Score': [],
'Test_R2_Score':[],
'Bias':[],
'Variance':[],
'Status':[]
}
max_rmse = 0.95
for reg_name,reg in regressors.items(): 
    reg.fit(x_train,y_train)
    predictions_train = reg.predict(x_train)
    predictions_test = reg. predict(x_test)
    mse_train = mean_squared_error(y_train, predictions_train)
    rmse_train = np.sqrt(mse_train)
    r2_train = r2_score(y_train, predictions_train)
    mse_test = mean_squared_error (y_test, predictions_test)
    rmse_test = np.sqrt(mse_test)
    r2_test = r2_score(y_test, predictions_test)
    result_dict[ 'Regressor'].append(reg_name)
    result_dict[ 'Train_MSE']. append(mse_train)
    result_dict['Train_R2_Score']. append(r2_train)
    result_dict[ 'Test_MSE']. append (mse_test)
    result_dict[ 'Test_R2_Score'].append(r2_test)
    bias = abs (0.95-r2_train)
    var = abs(r2_train - r2_test)
    result_dict[ 'Bias']. append (bias)
    result_dict[ 'Variance'].append(var)
    if bias > 0.1 and var > 0.1:
        status = "worst"
    elif bias > 0.1 and var < 0.1:
        status = "Underfitting"
    elif bias < 0.1 and var > 0.1:
        status = "Overfitting"
    elif bias < 0.1 and var < 0.1:
        status = "Best"
    result_dict['Status'].append(status)
Report_df = pd.DataFrame(result_dict)
print( "Final Report")
print(tabulate(Report_df, headers='keys', tablefmt='plane'))





```
